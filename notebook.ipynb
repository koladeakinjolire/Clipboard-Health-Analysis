```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Imported the Dataset
df = pd.read_csv('data.csv')
#Dropped the Unnamed Columns
df = df.drop(df.columns[0], axis=1)
#Created a total column
df['Total'] = df['PAY'] * df['ACCEPTED']
#Quick summary statistic of columns
print(df.head())
print(df.describe())
print(df['Total'].sum())
i = 0
total_list = []
while i < 12:   
	#Poisson distribution of proposed $30 per rider for 1000 rides in the first month
	Pay = np.random.poisson(30,1000)
    #Poisson distribution of proposed acceptance of 1000 rides with lambda of 0.55
	Accepted = np.random.poisson(0.527,1000)
	df2 = pd.DataFrame()
	df2['Pay'] = Pay
	df2['Accepted'] = Accepted
	df2['Total'] = df2['Pay'] * df2['Accepted']
	expense = df2['Total'].sum()
	total_list.append(expense)
	i+=1
print (df2.head())
print(df2.describe())
print(sum(total_list))
```
